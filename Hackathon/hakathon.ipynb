## Preprocessing data

```python
def impute_missing_values(df):
    # Create imputers
    numeric_imputer = SimpleImputer(strategy='median')
    categorical_imputer = SimpleImputer(strategy='most_frequent')
    
    # Identify numeric and categorical columns
    numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
    categorical_cols = df.select_dtypes(include=['object']).columns
    
    # Impute numeric columns
    df[numeric_cols] = numeric_imputer.fit_transform(df[numeric_cols])
    
    # Impute categorical columns
    df[categorical_cols] = categorical_imputer.fit_transform(df[categorical_cols])
    
    return df
```

```python
def merge_dataframes(df1, df2, key_column):
    """
    Merges two DataFrames based on a common key column.
    
    Parameters:
    - df1: First DataFrame
    - df2: Second DataFrame
    - key_column: The column on which to merge the DataFrames
    
    Returns:
    - Merged DataFrame
    """
    
    merged_df = pd.merge(df1, df2, on=key_column, how='inner')
    return merged_df
```

## Feature Engineering

```python
def engineer_features(df):
    # Creating a copy of the dataframe to avoid modifying the original one
    df_engineered = df.copy()
    
    # Average Delay
    df_engineered['Average_Delay'] = (df_engineered['Departure_Delay_in_Mins'] + df_engineered['Arrival_Delay_in_Mins']) / 2.0
    
    # Total Service Quality
    service_columns = ["Seat_Comfort", "Catering", "Onboard_Wifi_Service", "Onboard_Entertainment",
                       "Online_Support", "Ease_of_Online_Booking", "Legroom", "Online_Boarding",
                       "Onboard_Service", "Baggage_Handling", "CheckIn_Service", "Cleanliness"]
    
    # Age Group
    bins_age = [0, 30, 60, float('inf')]
    labels_age = ["Young", "Middle-aged", "Senior"]
    df_engineered['Age_Group'] = pd.cut(df_engineered['Age'], bins=bins_age, labels=labels_age, right=False)
    
    # Distance Group
    bins_distance = [0, 1000, 2000, float('inf')]
    labels_distance = ["Short", "Medium", "Long"]
    df_engineered['Distance_Group'] = pd.cut(df_engineered['Travel_Distance'], bins=bins_distance, labels=labels_distance, right=False)
    
    # Is_Delayed
    df_engineered['Is_Delayed'] = ((df_engineered['Departure_Delay_in_Mins'] > 0) | (df_engineered['Arrival_Delay_in_Mins'] > 0)).astype(int)
    
    # Total_Delay
    df_engineered['Total_Delay'] = df_engineered['Departure_Delay_in_Mins'] + df_engineered['Arrival_Delay_in_Mins']
    
    return df_engineered
```

```python
def encode_data_updated(df):
    # Creating a copy of the dataframe to avoid modifying the original one
    encoded_df = df.copy()
    
    # Ordinal Encoding
    ordinal_columns = [
        "Seat_Comfort", "Arrival_Time_Convenient", "Catering", "Onboard_Wifi_Service", 
        "Onboard_Entertainment", "Ease_of_Online_Booking", "Legroom", "Online_Boarding", 
        "Platform_Location", "Online_Support", "Onboard_Service", "Baggage_Handling", 
        "CheckIn_Service", "Cleanliness"
    ]
    
    # Defining the order for ordinal columns based on common values
    order = [
        ["Extremely Poor", "Poor", "Needs Improvement", "Acceptable", "Good", "Excellent"],
        ['Very Inconvenient', "Inconvenient", "Needs Improvement", "Manageable", "Convenient", "Very Convenient"]
    ]
    
    for col in ordinal_columns:
        if col in ["Platform_Location"]:
            ordinal_encoder = OrdinalEncoder(categories=[order[1]])
        else:
            ordinal_encoder = OrdinalEncoder(categories=[order[0]])
        encoded_df[col] = ordinal_encoder.fit_transform(encoded_df[[col]])
    
    # Label Encoding
    label_columns = ["Seat_Class", "Gender", "Customer_Type", "Type_Travel", "Travel_Class", "Age_Group", "Distance_Group"]
    label_encoder = LabelEncoder()
    for col in label_columns:
        encoded_df[col] = label_encoder.fit_transform(encoded_df[col].astype(str))
    
    return encoded_df
```

```python
def encode_specific_columns(train_df, test_df):
    # Creating a copy of the dataframes to avoid modifying the original ones
    encoded_train = train_df.copy()
    encoded_test = test_df.copy()

    # Columns to be encoded
    columns_to_encode = ["Age_Group", "Distance_Group"]

    for col in columns_to_encode:
        # Using LabelEncoder to encode the column
        label_encoder = LabelEncoder()
        
        # Fit the encoder using the training data and transform both training and testing data
        encoded_train[col] = label_encoder.fit_transform(encoded_train[col])
        encoded_test[col] = label_encoder.transform(encoded_test[col])

    return encoded_train, encoded_test
```

## Normalization

```python
encoded_train_data = pd.read_csv('encoded_train_data.csv')
encoded_test_data = pd.read_csv('encoded_test_data.csv')
```

```python
# Columns to exclude from normalization
cols_to_exclude_train = ['ID', 'Overall_Experience']
cols_to_exclude_test = ['ID']

# Check for columns in train data but not in test data
missing_cols = set(encoded_train_data.columns) - set(encoded_test_data.columns) - set(cols_to_exclude_train)
for col in missing_cols:
    encoded_test_data[col] = 0  # Add the missing column to the test data and fill with zeros

# Separate out the columns that shouldn't be normalized
train_data_to_exclude = encoded_train_data[cols_to_exclude_train]
test_data_to_exclude = encoded_test_data[cols_to_exclude_test]

# Drop the columns that shouldn't be normalized
encoded_train_data = encoded_train_data.drop(columns=cols_to_exclude_train)
encoded_test_data = encoded_test_data.drop(columns=cols_to_exclude_test)

# Ensure the order of columns in the test data is the same as in train data
encoded_test_data = encoded_test_data[encoded_train_data.columns]

# Initialize the scaler
scaler = StandardScaler()

# Normalize the data
encoded_train_data_normalized = pd.DataFrame(scaler.fit_transform(encoded_train_data), columns=encoded_train_data.columns)
encoded_test_data_normalized = pd.DataFrame(scaler.transform(encoded_test_data), columns=encoded_test_data.columns)

# Add back the columns that shouldn't be normalized
encoded_train_data_normalized = pd.concat([train_data_to_exclude, encoded_train_data_normalized], axis=1)
encoded_test_data_normalized = pd.concat([test_data_to_exclude, encoded_test_data_normalized], axis=1)

# Drop the missing columns from the normalized test data
encoded_test_data_normalized = encoded_test_data_normalized.drop(columns=missing_cols)

# Save the normalized data to new CSV files
encoded_train_data_normalized.to_csv('encoded_train_data_normalized.csv', index=False)
encoded_test_data_normalized.to_csv('encoded_test_data_normalized.csv', index=False)
```

## Model Pipeline Script 

```python
from sklearn.metrics import accuracy_score, classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
from sklearn.model_selection import RandomizedSearchCV
import numpy as np
import joblib

```

```python
encoded_train_data = pd.read_csv('encoded_train_data_normalized.csv')
encoded_test_data = pd.read_csv('encoded_test_data_normalized.csv')
# Drop the 'ID' column and separate the features and target variable for training data
X_train = encoded_train_data.drop(columns=['ID', 'Overall_Experience'])
y_train = encoded_train_data['Overall_Experience']

# Drop the 'ID' column and separate the features and target variable for testing data
X_test = encoded_test_data.drop(columns=['ID'])
```

```python
def randomized_search(classifiers, param_distributions, X_train, y_train):
    best_classifiers = {}
    for name, clf in classifiers:
        print(f'------CLASSIFIER: {name} STARTING-------')
        search = RandomizedSearchCV(clf, param_distributions=param_distributions[name],
                                    n_iter=50, scoring='accuracy', n_jobs=-1, cv=5, random_state=0)
        print('Fitting')
        search.fit(X_train, y_train)
        print('Fit')
        
        # Save the best model for this classifier
        best_model_path = f"{name.replace(' ', '_')}_best_model.pkl"
        joblib.dump(search.best_estimator_, best_model_path)
        print(f'Saved best {name} model to {best_model_path}')
        
        best_classifiers[name] = search.best_estimator_
        print(f'------CLASSIFIER: {name} ENDING-------')
    return best_classifiers

def get_best_classifier_from_search(best_classifiers, X_train, y_train):
    print('Choosing best classifier')
    scores = {}
    for name, clf in best_classifiers.items():
        scores[name] = accuracy_score(y_train, clf.predict(X_train))
    best_classifier_name = max(scores, key=scores.get)
    best_classifier = best_classifiers[best_classifier_name]
    
    # Save the overall best classifier
    best_model_path = f"Overall_best_{best_classifier_name.replace(' ', '_')}.pkl"
    joblib.dump(best_classifier, best_model_path)
    print(f'Saved overall best model to {best_model_path}')
    
    return best_classifier_name, best_classifier

def predict_and_save_best_model(best_classifier, X_test, encoded_test_data):
    print('Saving Data')
    # Predict on the X_test data
    y_pred = best_classifier.predict(X_test)
    
    # Convert the 'float_column' to an integer
    encoded_test_data['ID'] = encoded_test_data['ID'].astype(int)

    # Create a new dataframe with 'ID' and 'Overall_Experience'
    predicted_df = pd.DataFrame({
        'ID': encoded_test_data['ID'],
        'Overall_Experience': y_pred
    })
    predicted_df.to_csv('Submission3.csv', index=False)
```

```python
param_distributions = {
    "XGBoost": {
        'learning_rate': [0.01, 0.1],
        'n_estimators': [50, 100],
        'max_depth': [3, 5, 7],
        'subsample': [0.8, 1.0],
        'colsample_bytree': [0.8, 1.0],
        'gamma': [0, 0.1, 0.2]
    }
}
```

```python
classifiers = [("XGBoost", xgb.XGBClassifier(random_state=0, use_label_encoder=False))]
```

```python
# Perform Randomized Search
best_classifiers = randomized_search(classifiers, param_distributions, X_train, y_train)

# Get the best classifier from the search results
best_classifier_name, best_classifier = get_best_classifier_from_search(best_classifiers, X_train, y_train)

# Predict and save results using the best classifier
predict_and_save_best_model(best_classifier, X_test, encoded_test_data)
```

```python
def load_model_and_predict(filename, X_test, encoded_test_data):
    """
    Load a saved model from a specified file and use it to make predictions on the provided data.
    
    Parameters:
    - filename: The name of the file from which to load the model.
    - X_test: The test data to make predictions on.
    - encoded_test_data: The encoded test data to construct the final prediction dataframe.
    
    Returns:
    - predicted_df: A dataframe with the predictions.
    """
    # Load the saved model
    loaded_model = joblib.load(filename)
    
    # Predict using the loaded model
    y_pred = loaded_model.predict(X_test)
    
    # Convert the 'float_column' to an integer
    encoded_test_data['ID'] = encoded_test_data['ID'].astype(int)

    # Create a dataframe with predictions
    predicted_df = pd.DataFrame({
        'ID': encoded_test_data['ID'],
        'Overall_Experience': y_pred
    })
    
    # Optionally, save to a CSV file
    predicted_df.to_csv('LoadedModel_Submission.csv', index=False)
    
    print(f"Predictions saved to LoadedModel_Submission.csv")
    
    return predicted_df
```
